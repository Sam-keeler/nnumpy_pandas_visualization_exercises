```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
df
```

```python
# Create a column named passing_english that indicates whether each student has a passing grade in english.

df.drop(columns=['Passing english'], inplace = True)
df['Passing_english'] = df.english >= 70
```

```python
df
```

```python
# Sort the english grades by the passing_english column. How are duplicates handled?
# Sorted by the index

df.sort_values(by = 'Passing_english')
```

```python
# Sort the english grades first by passing_english and then by student name.

df.sort_values(by = ['Passing_english', 'name'])
```

```python
# Sort the english grades first by passing_english, and then by the actual english grade.

df.sort_values(by = ['Passing_english', 'english'])
```

```python
# Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.

df['overall'] = (df.math + df.english + df.reading)/3
```

```python
mpg = data('mpg')
mpg
```

```python
# How many rows and columns are there?
# 11 columns
# 234 rows

print(mpg.info())
```

```python
# What are the data types of each column?

mpg.dtypes
```

```python
# Summarize the dataframe with .info and .describe

print(mpg.info())
print(mpg.describe())
```

```python
# Rename the cty column to city.

mpg.rename(columns = {'cty' : 'city'}, inplace = True)
```

```python
# Rename the hwy column to highway.

mpg.rename(columns = {'hwy' : 'highway'}, inplace = True)
```

```python
# Do any cars have better city mileage than highway mileage?
# No

mpg[mpg.city > mpg.highway]
```

```python
# Create a column named mileage_difference.
# This column should contain the difference between highway and city mileage for each car.

mpg['mileage_difference'] = (mpg.highway - mpg.city)
```

```python
# Which car (or cars) has the highest mileage difference?
# 1999 VW Beetle, 2008 Honda Civic

mpg.nlargest(1, 'mileage_difference', keep='all')
```

```python
# Which compact class car has the lowest highway mileage?
# 1999 VW Jetta auto(I4)

mpg[mpg['class'] == 'compact'].sort_values(by = 'highway').head()
```

```python
# Which compact class car has the highest highway mileage?
# 1999 VW Jetta manual(M5)

mpg[mpg['class'] == 'compact'].sort_values(by = 'highway', ascending = False).head()
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.

mpg['average_mileage'] = (mpg.highway + mpg.city)/2
```

```python
mpg.head()
```

```python
# Which dodge car has the worst average mileage?
# 2008 dodge ram 1500

mpg[mpg.manufacturer == 'dodge'].sort_values(by = 'average_mileage').head(1)
```

```python
# Which dodge car has the best average mileage?
# 1999 dodge caravan

mpg[mpg.manufacturer == 'dodge'].sort_values(by = 'average_mileage', ascending = False).head(1)
```

```python
len(mammals[mammals.hoppers == True])
```
