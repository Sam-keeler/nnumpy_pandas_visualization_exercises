```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
mpg = data('mpg')
```

```python
mpg['avg_mpg'] = (mpg.hwy + mpg.cty)/2
```

```python
mpg
```

```python
# On average, which manufacturer has the best miles per gallon?
# Honda

mpg.groupby('manufacturer').avg_mpg.mean().sort_values(ascending = False).head(1)
```

```python
# How many different manufacturers are there?
# 15

mpg.groupby('manufacturer').avg_mpg.mean().count()
```

```python
# How many different models are there?
#38

mpg.groupby('model').avg_mpg.mean().count()
```

```python
mpg['a_or_m'] = mpg['trans'].str[0]
```

```python
mpg
```

```python
# Do automatic or manual cars have better miles per gallon?
# manual

mpg.groupby('a_or_m').avg_mpg.mean()
```

```python
# Create a function named get_db_url. It should accept a username, hostname, password, and database 
# name and return a url formatted like in the examples in this lesson.

from env import host, user, password

def get_db_url(db, host = host, user = user, password = password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
```

```python
# Use your function to obtain a connection to the employees database.

pd.read_sql('SELECT * FROM employees', get_db_url('employees'))
```

```python
# Read the employees and titles tables into two separate dataframes

employees = pd.read_sql('SELECT * FROM employees', get_db_url('employees'))
titles = pd.read_sql('SELECT * FROM titles', get_db_url('employees'))
```

```python
# Visualize the number of employees with each title.

titles.groupby('title').emp_no.agg('count').plot.bar()
```

```python
# Join the employees and titles dataframes together.

t_and_e = pd.merge(titles, employees, left_on='emp_no', right_on='emp_no')
```

```python
# Visualize how frequently employees change titles.

t_and_e.groupby('emp_no').title.agg('count').head().plot.bar()
```

```python
# For each title, find the hire date of the employee that was hired most recently with that title.

t_and_e.groupby('title').hire_date.max()
```

```python
chips = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
```

```python
def fixit(x):
    x = x.replace('$', '')
    x = x.replace(',', '')
    x = float(x)
    return x
```

```python
fixit('$806.73')
```

```python
chips['item_price'] = chips.item_price.apply(lambda s: fixit(s))
chips
```

```python
# What is the total price for each order?

chips.groupby('order_id').item_price.sum()
```

```python
# What are the most popular 3 items?
# chicken bowl, chicken burrito, chips and guac

chips.groupby("item_name").agg({"quantity": np.sum}).sort_values(by = 'quantity', ascending = False).head(3)
```

```python
# Which item has produced the most revenue?
# chicken bowl

chips.groupby("item_name").agg({"item_price": np.sum}).sort_values(by = 'item_price', ascending = False).head(1)
```
