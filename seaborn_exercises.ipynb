```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# What does the distribution of petal lengths look like?

sns.displot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?
# Yes

sns.relplot(data = iris, x = 'petal_length', y = 'petal_width')
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?
# For Setosa at least, yes

sns.relplot(data = iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
```

```python
# Which features would be best used to predict species?
# Petal width looks like the best indicator

plt.figure(figsize=(12, 12))
plt.suptitle('Fun with Seaborn')

plt.subplot(221)
sns.boxplot(data = iris, x = 'sepal_length', y = 'species')
plt.title('Sepal Length')

plt.subplot(222)
sns.boxplot(data = iris, x = 'sepal_width', y = 'species')
plt.title('Sepal Width')

plt.subplot(223)
sns.boxplot(data = iris, x = 'petal_width', y = 'species')
plt.title('Petal Width')

plt.subplot(224)
sns.boxplot(data = iris, x = 'petal_length', y = 'species')
plt.title('Petal Length')
```

```python
swiss = data('swiss')
swiss
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)
swiss['is_catholic'] = (swiss['Catholic'] > 50)
```

```python
swiss
```

```python
# Does whether or not a province is Catholic influence fertility?
# It seems to influence fertility a significant amount

sns.boxplot(data = swiss, x = 'is_catholic', y = 'Fertility')
```

```python
# What measure correlates most strongly with fertility?
# Looks like education

from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2

plt.figure(figsize=(12, 18))
plt.suptitle('Fertility Correlations')

plt.subplot(511)
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Agriculture')

plt.subplot(512)
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Examination')

plt.subplot(513)
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Education')

plt.subplot(514)
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Catholic')

plt.subplot(515)
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Infant.Mortality')
```

```python
from env import host, user, password

def get_db_url(db, host = host, user = user, password = password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'

chips = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))

def fixit(x):
    x = x.replace('$', '')
    x = x.replace(',', '')
    x = float(x)
    return x

chips['item_price'] = chips.item_price.apply(lambda s: fixit(s))
chips
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
chips = chips.assign(total_revenue=chips.groupby('item_name').item_price.transform('sum'))
chips
```

```python
sns.barplot(data = chips, x = 'item_name', y = 'total_revenue', order = ['Chicken Bowl', 'Chicken Burrito', 'Steak Burrito', 'Steak Bowl'])
```
