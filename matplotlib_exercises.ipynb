```python
import math
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
y = [x**2 - x + 2 for x in x]
plt.plot(x, y)
plt.annotate('Origin', xy=(0, 0), xytext=(0, 10),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
n_rows = 2
n_cols = 2
plt.figure(figsize=(14, 8))

x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
a = [math.sqrt(x) for x in x]
b = [x**3 for x in x]
c = [2**x for x in x]
d = [(1/(x+1)) for x in x]

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, a)
plt.title(r'$\sqrt{x}$')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, b)
plt.title('$x^3$')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, c)
plt.title('$2^x$')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, d)
plt.title('1 / (x +1)')

plt.show()
```

```python
x = [0, 1, 2, 3, 4]
a = [math.sqrt(x) for x in x]
b = [x**3 for x in x]
c = [2**x for x in x]
d = [(1/(x+1)) for x in x]


plt.plot(x, a, c='orange', label=r'$\sqrt{x}$')
plt.plot(x, b, c='green', label='$x^3$')
plt.plot(x, c, c='blue', label='$2^x$')
plt.plot(x, d, c='black', label='1 / (x +1)')

plt.legend(loc='upper left')
plt.xlabel('$x$')

plt.show()
```
