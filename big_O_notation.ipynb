```python
import matplotlib.pyplot as plt
import math
```

```python
y = range(1, 4)
x = range(1, 31)
x = [x/10 for x in x]
a = [1 for x in x]
b = [math.log(x) for x in x]
c = [x * math.log(x) for x in x]
d = [x**2 for x in x]
e = [2**x for x in x]
f = [math.factorial(y) for y in y]
g = [x**x for x in x]

plt.figure(figsize=(14, 10))

plt.plot(x, a, c='orange', label='O(1)')
plt.plot(x, b, c='green', label='O(log n)')
plt.plot(x, x, c='blue', label ='O(n)')
plt.plot(x, c, c='black', label='O(n log n)')
plt.plot(x, d, c='red', label='O($n^2$)')
plt.plot(x, e, c='yellow', label='O($2^n$)')
plt.plot(y, f, c='purple', label='O(n!)')
plt.plot(x, g, c='cyan', label='O($n^n$)')

plt.legend(loc='upper left')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(-0.3, 28, 'Sam Keeler', fontsize=10, color='blue', bbox=props)

plt.show()
```
